```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df_hul_reg = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="hul")
df_hind_beta = []
df_dates = ['2020_2019','2021_2020','2022_2021','2023_2022']
df_hind_2020_2019 = df_hul_reg[(df_hul_reg['Date'] >= '2019-01-01') & (df_hul_reg['Date'] <= '2020-01-01')]
df_hind_2021_2020 = df_hul_reg[(df_hul_reg['Date'] >= '2020-01-01') & (df_hul_reg['Date'] <= '2021-01-01')]
df_hind_2022_2021 = df_hul_reg[(df_hul_reg['Date'] >= '2021-01-01') & (df_hul_reg['Date'] <= '2022-01-01')]
df_hind_2023_2022 = df_hul_reg[(df_hul_reg['Date'] >= '2022-01-01') & (df_hul_reg['Date'] <= '2023-01-01')]
print(df_hind_2020_2019.head())

df_nifty50 = pd.read_excel("D:/IT/dav/stocks.xlsx")
df_nifty50_2020_2019 = df_nifty50[(df_nifty50['Date'] >= '2019-01-01') & (df_nifty50['Date'] <= '2020-01-01')]
df_nifty50_2021_2020 = df_nifty50[(df_nifty50['Date'] >= '2020-01-01') & (df_nifty50['Date'] <= '2021-01-01')]
df_nifty50_2022_2021 = df_nifty50[(df_nifty50['Date'] >= '2021-01-01') & (df_nifty50['Date'] <= '2022-01-01')]
df_nifty50_2023_2022 = df_nifty50[(df_nifty50['Date'] >= '2022-01-01') & (df_nifty50['Date'] <= '2023-01-01')]
print(df_nifty50_2020_2019.head())

model = LinearRegression()
model.fit(df_hind_2020_2019[['percent_change']], df_nifty50_2020_2019['percent_change'])
c1 = model.coef_[0]
df_hind_beta.append(c1)

model.fit(df_hind_2021_2020[['percent_change']], df_nifty50_2021_2020['percent_change'])
c2 = model.coef_[0]
df_hind_beta.append(c2)

model.fit(df_hind_2022_2021[['percent_change']], df_nifty50_2022_2021['percent_change'])
c3 = model.coef_[0]
df_hind_beta.append(c3)

model.fit(df_hind_2023_2022[['percent_change']], df_nifty50_2023_2022['percent_change'])
c4 = model.coef_[0]
df_hind_beta.append(c4)

# Fit the model to the data
print(df_hind_beta)
plt.plot(df_dates,df_hind_beta)
plt.title("Beta")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_fmcg1 = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="fmcg_cnx")
df_pharma1 = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="pharma_cnx")
df_metal1 = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="metal_cnx")
df_it1 = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="it_cnx")
df_auto1 = pd.read_excel("D:/IT/dav/Stocks_Data.xlsx",sheet_name="auto_cnx")

fmcg = df_fmcg1[(df_fmcg1['Date'] >= '2020-02-10') & (df_fmcg1['Date'] <= '2020-04-20')]
pharma = df_pharma1[(df_pharma1['Date'] >= '2020-02-10') & (df_pharma1['Date'] <= '2020-04-20')]
metal = df_metal1[(df_metal1['Date'] >= '2020-02-10') & (df_metal1['Date'] <= '2020-04-20')]
it= df_it1[(df_it1['Date'] >= '2020-02-10') & (df_it1['Date'] <= '2020-04-20')]
auto = df_auto1[(df_auto1['Date'] >= '2020-02-10') & (df_auto1['Date'] <= '2020-04-20')]

fig,axs = plt.subplots(5,figsize=(5,15))
axs[0].plot(fmcg['Date'],fmcg['Close'])
axs[1].plot(pharma['Date'],pharma['Price'])
axs[2].plot(metal['Date'],metal['Close'])
axs[3].plot(it['Date'],it['Close'])
axs[4].plot(auto['Date'],auto['Close'])
axs[0].title.set_text("FMCG")
axs[1].title.set_text("PHARMA")
axs[2].title.set_text("METAL")
axs[3].title.set_text("IT")
axs[4].title.set_text("AUTO")
plt.setp(axs[0].get_xticklabels(), rotation=45)
plt.setp(axs[1].get_xticklabels(), rotation=45)
plt.setp(axs[2].get_xticklabels(), rotation=45)
plt.setp(axs[3].get_xticklabels(), rotation=45)
plt.setp(axs[4].get_xticklabels(), rotation=45)
fig.tight_layout()
```
